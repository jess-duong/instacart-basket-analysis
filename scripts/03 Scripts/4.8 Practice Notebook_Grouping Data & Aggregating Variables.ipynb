```python
# Group df by column
df.groupby('product_name')
```

```python
# Single aggregation: Mean order_number by department
# Group the data by department_id and calculate
# the mean number of orders per user (order_number)

dept_order_mean = df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Dispay the result
dept_order_mean
```

```python
# Single aggregation using mean() instead of agg()
#This calculates the same result as the agg() example, but uses the built-in mean() function directly
dept_order_mean_alt = df.groupby('department_id')['order_number'].mean()
```

```python
# Display the result
dept_order_mean_alt
```

```python
# Multiple aggregations using agg()
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# Create Loyalty flag using loc

```python
# Creating max_order using groupby + transform
# This calculates the maximum order_number for each used
# and assigns that value back to every row for that user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# Create loyalty_flag based on max_order thresholds

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Confirming flag was created correctly
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
