# Instacart Basket Analysis

This notebook executes the following project tasks:
1. Import the cleaned Instacart orders dataset (`orders_checked.csv`)
2. Import the prior order–product dataset (`orders_products_prior.csv`)
3. Merge both datasets using the shared `order_id` key
4. Validate the merge using the `_merge` indicator column
5. Check the resulting dataframe shape to confirm expected row volume
6. Export the combined dataset as a Pickle file for efficient storage (`orders_products_combined.pkl`)

All work performed here directly contributes to building the consolidated Instacart dataset for downstream analysis and segmentation.

```python
#Import libraries
import pandas as pd
import numpy as nb
import os
```

```python
#02. Setting main project path
#Base path to the Instacart project folder
path = r'/Users/jessduong/Documents/CF/Achievement 4_Python/12-2025 Instacart Basket Analysis'
```

```python
#Import orders_products_prior (new dataset)
df_ords_prior = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'orders_products_prior.csv'))
```

```python
#Import latest, cleanest orders data set
df_ords = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_checked.csv'))
```

```python
#Check the output
df_ords_prior.head()
```

```python
#Check the output
df_ords.head()
```

```python
#Check shape of new data set
df_ords_prior.shape
```

```python
#Check shape of cleaned dataset (checked)
df_ords.shape
```

```python
#Merging datasets
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
#Check the output
df_merged_large.head()
```

```python
df_merged_large['_merge'].value_counts()
```

```python
#Outer merge diagnostic check to verify if true match (note overwriting)
pd.merge(df_ords, df_ords_prior, on = 'order_id', how = 'outer', indicator = True)['_merge'].value_counts()
```

```python
#Export to CV
df_merged_large.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.csv'), index=False)
```

```python
#Step 2: Export merged file in pickle format as "orders_products_combined.pkl."
df_merged_large.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```
